# Problem Definition - House Price Prediction


## Problem Statement:
The goal of this project is to predict the selling price of a house based on features like size, number of bedrooms, location, and quality.

## Input and Output
- **Input (X):** LotArea, OverallQual, YearBuilt, etc.  
- **Output (y):** SalePrice (continuous numeric value)
